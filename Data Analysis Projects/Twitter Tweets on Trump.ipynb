# Twitter Tweets on Ex-President Donald Trump

As a political figure, Donald J. Trump used Twitter to praise, to cajole, to entertain, to lobby, to establish his version of events — and, perhaps most notably, to amplify his scorn. This list documents the verbal attacks Mr. Trump posted on Twitter, from when he declared his candidacy in June 2015 to Jan. 8, when Twitter permanently barred him.

Based on an analysis of tweets since Mr. Trump declared his candidacy for president, on June 16, 2015. Retweets are not included. Some names may be omitted. This was first published in 2016; some of the people insulted have since died, and some people’s titles or public roles have changed.

Source of Dataset:

This dataset has been extracted from The New York Times.

# Importing Packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import STOPWORDS, ImageColorGenerator, WordCloud
from PIL import Image
```

```python
df = pd.read_csv("trump_insult_tweets_2014_to_2021.csv")
df.head()
```

# Building a repeated words in Tweets using WordCloud

```python
tweet_text = " ".join(df["tweet"])
wordcloud_tweet = WordCloud(width=1500, height=500).generate(tweet_text)
plt.figure(figsize=(30,20))
plt.imshow(wordcloud_tweet, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
df.head()
```

# Building a repeated words in Insult texts using WordCloud

```python
insult_text = " ".join(df["insult"])
wordcloud_insult = WordCloud(width= 1500, height=500).generate(insult_text)
plt.figure(figsize=(30,20))
plt.axis("off")
plt.imshow(wordcloud_insult, interpolation="bilinear")
plt.show()
```

```python
df.head()
```

# Building a repeated words in Target Text using WordCloud

```python
target_text = " ".join(df["target"])
wordcloud_target = WordCloud(width=1500, height=500).generate(target_text)
plt.figure(figsize=(30,20))
plt.imshow(wordcloud_target, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
df.head()
```

```python
df["Tweet Length"] = df["tweet"].apply(len)
```

```python
df.tail()
```

```python
df.head()
```

```python
df["Year"] = pd.to_datetime(df["date"])
```

```python
df["Year"] = pd.to_datetime(df["date"]).dt.year
df["Month"] = pd.to_datetime(df["date"]).dt.month
df["Day"] = pd.to_datetime(df["date"]).dt.day
```

```python
df.head()
```

```python
df.shape
```

```python
df["tweet"] = df["tweet"].replace({"\$" : ""}, regex=True)
```

```python
df.head()
```

# Count of Tweets through the years 2014-2021

```python
plt.xlabel("Years")
plt.ylabel("Count of tweets")
plt.title("Tweets count by Years")
df.groupby(["Year"])["tweet"].agg("count").plot(kind= "bar")
```

# Average trend of Tweet counts on Month Basis

```python
plt.xlabel("Months")
plt.ylabel("Count of tweets")
plt.title("Tweets count by Months")
df.groupby(["Month"])["tweet"].agg("count").plot(kind= "bar")
```

```python
year_2020 = df[df["Year"] == 2020]
```

# Count of Tweets in the year 2020

```python
plt.xlabel("Months")
plt.ylabel("Count of tweets")
plt.title("Tweets count by Months in 2020")
year_2020.groupby(["Month"])["tweet"].agg("count").plot(kind= "bar")
```

As observed, Highest number of tweets have been recorded in the year 2020 which could probably be because of the Pandemic and the way Mr. President Donald Trump has handled the on-going issues. 

Due to the spread of Trump's hatred tweets, Trump's Twitter Account has been barred permanently.
